```python
import numpy as np
from pathlib import Path
import torch
import matplotlib.pyplot as plt
import torch.nn.functional as F
```

```python
labels_npz = np.load(Path("dataset") / "annotations_processed.npz")
num_labels = 33
```

```python
all_video_label_count = []
for fn in labels_npz.files:
    label_count = F.one_hot(torch.from_numpy(labels_npz[fn]), num_labels).sum(dim=0)
    all_video_label_count.append(label_count)

plt.bar(x=np.arange(num_labels), height=torch.stack(all_video_label_count).sum(dim=0).numpy())
plt.show()
```
